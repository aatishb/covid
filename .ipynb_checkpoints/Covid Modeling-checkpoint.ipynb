```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
```

```python
df
```

```python
summary = df.iloc[:,[1,-1]].groupby('Country/Region').sum()
mostrecentdate = summary.columns[0]
summary.sort_values(by=mostrecentdate, ascending = False)
```

```python
def plotCases(country):
    sk = df[df['Country/Region'] == country].iloc[:,4:].T.sum(axis = 1)
    sk.columns = ['Cases']
    sk.plot(legend = False, title = country + ' COVID-19 Cases')
    plt.show()
```

```python
plotCases('South Korea')
```

```python
plotCases('Italy')
```

```python
plotCases('Iran')
```

```python
plotCases('Japan')
```

```python
plotCases('US')
```
